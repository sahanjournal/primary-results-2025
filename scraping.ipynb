```python
import pandas as pd
import urllib.request
import requests
import re 
```

```python
# update url from state election results website

url = "https://electionresultsfiles.sos.mn.gov/20250812/local.txt"
```

```python
# read in data in .txt format

response = requests.get(url)
data = response.text
data = data.replace('\r', '')
data = data.split('\n')

data
```

```python
# save as .txt format
with open('results.txt', 'w') as f:
    for line in data:
        f.write(line)
        f.write('\n')
```

```python
# open in pandas as csv format

df = pd.read_csv("results.txt",delimiter = ';',header=None,)
df.head()
```

```python
# add column names

df.columns = [
    'State',
	'County ID (if applicable)',
	'Precinct name (if applicable)',
	'Office ID',
	'Office Name', 
	'District*',
	'Candidate Order Code',
	'Candidate Name (First/Last/Suffix all in one field)',
	'Suffix (not used)',
	'Incumbent Code (not used)',
	'Party Abbreviation',
	'Number of Precincts reporting',
	'Total number of precincts voting for the office',
	'Votes for Candidate',
	'Percentage of Votes for Candidate out of Total Votes for Office',
	'Total number of votes for Office in area'
]
```

```python
# MAKE SOME CHANGES TO THE COLUMN NAMES

df['Candidate Name (First/Last/Suffix all in one field)'] = df['Candidate Name (First/Last/Suffix all in one field)'].str.replace("Write-In","Write-in")


df.rename(columns={"Candidate Name (First/Last/Suffix all in one field)": "Candidate", "Votes for Candidate": "Votes", "Percentage of Votes for Candidate out of Total Votes for Office": "Percentage"}, inplace=True)
```

```python
house_url = "https://electionresultsfiles.sos.mn.gov/20250812_1/LegislativeByDistrict.txt"
# read in data in .txt format

house_response = requests.get(house_url)
house_data = house_response.text
house_data = house_data.replace('\r', '')
house_data = house_data.split('\n')

house_data
```

```python
# save as .txt format
with open('house_results.txt', 'w') as f:
    for line in house_data:
        f.write(line)
        f.write('\n')
```

```python
# open in pandas as csv format

house_df = pd.read_csv("house_results.txt",delimiter = ';',header=None,)
house_df.head()
```

```python
house_df.columns = [
    'State',
	'County ID (if applicable)',
	'Precinct name (if applicable)',
	'Office ID',
	'Office Name', 
	'District*',
	'Candidate Order Code',
	'Candidate',
	'Suffix (not used)',
	'Incumbent Code (not used)',
	'Party Abbreviation',
	'Number of Precincts reporting',
	'Total number of precincts voting for the office',
	'Votes',
	'Percentage',
	'Total number of votes for Office in area'
]
```

```python
house_df
```

```python
house_df = house_df[['Candidate','Percentage','Votes','Party Abbreviation','Number of Precincts reporting','Total number of precincts voting for the office']]
house_df
```

```python
# do not include R candidates
house_df = house_df[~house_df['Party Abbreviation'].isin(['R', 'RF'])]
house_df
```

```python
# save house file
house_df.to_csv("data/house_results.csv",index=False)
```
